```python
import numpy as np

# Replace "Varies with device" with NaN
cols_to_clean = ["Size", "Android Ver"]
data[cols_to_clean] = data[cols_to_clean].replace("Varies with device", np.nan)

# ----- Size -----
data["Size_raw"] = data["Size"]  # keep raw copy


# Convert KB to MB (optional)
def size_to_mb(size):
    if pd.isna(size) or size == "Varies with device":
        return np.nan
    size = size.strip()
    if size.endswith("M"):
        return round(float(size.replace("M", "")), 2)
    if size.endswith("k"):
        return round(float(size.replace("k", "")) / 1024, 2)
    return np.nan


data["Size(in MBs)"] = data["Size"].apply(size_to_mb)




# ----- Android Ver -----
data["Android_Ver_raw"] = data["Android Ver"]  # keep raw copy
data["Android Ver"] = (
    data["Android Ver"].astype(str)
                       .str.extract(r"(\d+(\.\d+)*)")[0]
)
data["Android Ver"] = pd.to_numeric(data["Android Ver"], errors="coerce")
```

These three columns contained a value "Varies with device". Keeping them would have created a lot of noise for Model Building or even for Analysis. So I replaced these values with NaN, so that I can drop them later.

Also the Column size contained the values in MB's as well as in Kb. I have converted all of them values to Kb's and converted the column to float type

Column Genre contains lot of unique values. It can create high cardinality issue. We will keep the less occuring Genres in a separate category "Others".

```python


keep_these = data['Genres'].value_counts()[data['Genres'].value_counts() >= 28].index 
data['Genres'] = data['Genres'].apply(lambda x : x if x in keep_these else "others")

data['Genres'].value_counts().count()
```

```python

others_ratio = (data['Genres'] == 'others').mean()

print(f"'Others' proportion: {others_ratio * 100 :.2f}")

```

Safe to Use!

We have kept the categories that had lesser number of counts in the Other's category. Previously we had 105 uniques categories, now we are left wtih only 42 and the proportion of the Others category only consists of 7% of the actual data which means it is safe to use...

```python
data['Content Rating'].value_counts()
```

In the Content Rating column, "Adults only 18+" is only two times in the dataset it is so low compared to other values in different categories that model will not learn anything from them or ignore them completely.
```python
fig, ax = plt.subplots(ncols=3, nrows=2, figsize=(30,25))
index=0
ax=ax.flatten()


for col,value in data.select_dtypes(include='number').drop('Rating', axis = 1).items():
    if index >= 24:
        break
    sns.boxenplot(value,ax=ax[index], color = 'brown')
    ax[index].set_title(col)
    index+=1
    
plt.tight_layout()
```

```python
fig, ax = plt.subplots(ncols=2, nrows=4, figsize=(30,25))
index=0
ax=ax.flatten()


for col,value in data.select_dtypes(include='number').items():
    if index >= 24:
        break
    sns.kdeplot(value,ax=ax[index], color='orangered', shade = True)
    ax[index].set_title(col)
    index+=1
    
plt.tight_layout()
```

We will do log transform before doing the outlier treatment as most of the outliers are on the extreme end and are making the graphs highly right skewed except the Rating column which is left skewed as it should be because most of the Ratings are between 3.5 and 5.

- **Note:** We will keep Android Version and Rating columns untouched as doing outlier treatment will mess up the values in the Android Version column and Ratings is our output column and doing anything with the column will make the predictions biased.

```python


def log_transformation(data, column):
    data[column] = np.log1p(data[column])
    return data[column]
```

```python
data.select_dtypes(include='number').columns
```

```python
outlier_columns = ['Reviews','Installs','Size(in MBs)','Days Since Last Update','Price in Dollars']
```

```python
data['Reviews'] = log_transformation(data,'Reviews')
data['Installs'] = log_transformation(data,'Installs')
data['Size(in MBs)'] = log_transformation(data,'Size(in MBs)')
data['Days Since Last Update'] = log_transformation(data,'Days Since Last Update')
data['Price in Dollars'] = log_transformation(data,'Price in Dollars')
```

```python
fig, ax = plt.subplots(ncols=2, nrows=4, figsize=(30,25))
index=0
ax=ax.flatten()


for col,value in data[outlier_columns].items():
    if index >= 24:
        break
    sns.kdeplot(value,ax=ax[index], color='forestgreen', shade = True)
    ax[index].set_title(col)
    index+=1
    
plt.tight_layout()
```

- Let's see the affect on outliers...

```python
fig, ax = plt.subplots(ncols=3, nrows=2, figsize=(30,25))
index=0
ax=ax.flatten()


for col,value in data[outlier_columns].items():
    if index >= 24:
        break
    sns.boxplot(value,ax=ax[index], color='lightgreen')
    ax[index].set_title(col)
    index+=1
    
plt.tight_layout()
```

- The Log Transformation reduced the impact of the outlier except "Price in Dollar" column as most of the outliers are fewer in number and at the farther side... we don't need to use Capping Method or Outlier Removal technique now to deal with the outliers except the "Price in Dollar" column.

- Price in Dollar has an extreme Outliers. It is obvious almost 70 - 80% of the data are free.

## Data Preprocessing

```python
data.head()
```

```python


X = data.drop('Rating', axis =1)

y = data['Rating']
```

```python


from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)

X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
data.select_dtypes(include='object')
```

```python


cat_cols = data.select_dtypes(include='object').columns
num_cols = data.select_dtypes(include='number').drop('Rating', axis = 1).columns
```

```python

from sklearn.preprocessing import OneHotEncoder, RobustScaler
from sklearn.compose import ColumnTransformer


oh_enc = OneHotEncoder(drop='first', sparse_output=False, dtype='int32')
scaler = RobustScaler()



preprocessing = ColumnTransformer(transformers=[
    ('onehot', oh_enc, cat_cols),
    ('scaler', scaler, num_cols)
])
```

```python
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.model_selection import cross_val_score

model = Pipeline(steps = [
    ('prep', preprocessing),
    ('lr', LinearRegression())
])


model.fit(X_train,y_train)

y_pred = model.predict(X_test)

y_pred_train = model.predict(X_train)



print("\n\nTraining Data Accuracy Score: ", r2_score(y_train,y_pred_train))
      
print("\n\nTesting Data Accuracy Score: ", r2_score(y_test,y_pred))
```

```python
from sklearn.tree import DecisionTreeRegressor


model = Pipeline(steps = [
    ('prep', preprocessing),
    ('lr', DecisionTreeRegressor())
])


model.fit(X_train,y_train)

y_pred = model.predict(X_test)

y_pred_train = model.predict(X_train)



print("\n\nTraining Data Accuracy Score: ", r2_score(y_train,y_pred_train))
      
print("\n\nTesting Data Accuracy Score: ", r2_score(y_test,y_pred))
```
